## las variables
1. CT --> costo del traje
2. DE --> Decuento que se obtendra
3. PF --> Precio final del traje
## formulas
* **DE = CT * (porcentaje_descuento / 100)**
* **PF = CT - DE**

## Diagrama de flujo lucidchart
<img src="https://raw.githubusercontent.com/VALERIAMARTINEZ128/Act-12-p3.2-determinar-el-pago-/refs/heads/main/DF%20DE%20CALCULO.png" width="300" height="500">

```python
costo_traje = 1000
porcentaje_descuento = 15
descuento = costo_traje * (porcentaje_descuento / 100)
precio_final = costo_traje - descuento
print(f"Costo del Traje (CT): ${costo_traje:.2f}")
print(f"Descuento (DE): ${descuento:.2f}")
print(f"Precio Final (PF): ${precio_final:.2f}")
```

```python
# para cuando es verdadero
# calculando el cambio uando if else
# datos de entrada CD,CP
# mostar cambio si es falso
# si es verdadero mostar No  alcanza
CP = float(input("ingrese el costo del producto: "))
CD = float(input("ingrese la cantidad de dinero: "))
if CD >= CP:
    cambio = CD - CP
    print("No alcanza")

else:
    cambio = CD - CP
    print("el cambio es: ",cambio)
    print("fin del programa")
    print("programa realizado por valeria yretzi NL = 32 ")
```
